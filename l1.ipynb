```python
# Q1.1
import numpy as np
rng = np.random.default_rng()

dice_result = rng.integers(1, 7, 100)
print(dice_result[:10])
```

```python
# Q1.2
mean = np.mean(dice_result)
dst = abs(mean - 3.5)
threshold = 0.1
print("close" if dst <= threshold else "not close")
```

```python
# Q2.1
size = 3
mtx = []
# 1
for _ in range(size):
    row = rng.integers(1, 10, size)
    mtx.append(row)
# 2
print(np.array(mtx))

# 3
identity = np.empty((size, size))
for i in range(size):
    for j in range(size):
        val = 0
        if i == j:
            val = 1
        identity[i][j] = val
print(identity)
```

```python
# Q2.2

# 1
transposed = np.empty((size, size))
for i in range(size):
    for j in range(size):
        transposed[j][i] = mtx[i][j]
print(transposed)

# 2
is_identical = True
for i in range(size):
    for j in range(size):
        if transposed[j][i] != mtx[i][j]:
            is_identical = False
print(is_identical)

```
