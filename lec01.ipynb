```python
import numpy as np

rolls = np.random.randint(1, 7, 100)

print("All rolls dice: ", rolls)
print ("The first 10 rolls dice: ", rolls[:10])
```

```python
mean_roll = np.mean(rolls)
print("Mean of rolls: ", mean_roll)

print("Expected value: ", 3.5)
print("Difference: ", mean_roll - 3.5)
```

```python
A = np.arange(1, 10).reshape(3, 3)
print("Matrix A:\n", A)
print("Shape of A:", A.shape)

I = np.eye(3)
print("Identity Matrix I:\n", I)
print("Shape of I:", I.shape)
```

```python
A_T = A.T
print("Transpose of A:\n", A_T)

i, j = 0, 1
is_equal = A[i, j] == A_T[j, i]
print(f"A[{i},{j}] == A_T[{j},{i}]:", is_equal)
```

```python
v = np.array([3, -4, 0])

l1_norm = np.sum(np.abs(v))
l2_norm = np.sqrt(np.sum(v**2))
linf_norm = np.max(np.abs(v))

print("Vector v:", v)
print("L1 norm of v:", l1_norm)
print("L2 norm of v:", l2_norm)
print("Linf norm of v:", linf_norm)
```
